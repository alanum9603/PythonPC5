#### Alicia quiere un departamento en Lisboa con habitaciones separadas, asimismo, solo va a aceptar lugares que tengan más de 10 críticas, además, que su puntuación sea mayor a 4, generamos una variable que contenga estas condiciones. Entonces, creamos un nuevo dataframe df_alicia que contenga el resultado de aplicar los filtros al dataframe df.
##### Los valores de las columnas reviews y overall_satisfaction son numéricos, por tanto, no se complicará la creación de las condicionales, para revisar el tipo de habitación existente en la columna room_type realizaremos lo siguiente:

```python
df['room_type'].unique()
```

#### A partir, de esto, se puede decidir que se usará el dato 'Private room'. Creames la condicional y guardamos el resultado en una nueva dataframe.

```python
condicion = (df['reviews'] > 10) & (df['overall_satisfaction'] > 4) & (df['room_type'] == 'Private room')
df_filtrado = df[condicion]
```

```python
df_filtrado.head()
```

```python
df_filtrado.tail()
```

#### Finalmente, ordenamos la lista para mostrar las mejores puntuaciones primero e ir descendiendo.

```python
df_ordenado = df_filtrado.sort_values(by=['overall_satisfaction','reviews'], ascending=False)
```
